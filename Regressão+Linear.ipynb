```python
import pandas as pd 
from sklearn.datasets import load_boston
```

```python
dataset = load_boston()
```

```python
dataset = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
dataset.head()
```

```python
print("RESPOSTA NUM. 2 E 3")
print(dataset.shape)
```

```python
print("RESPOSTA NUM. 5")
import seaborn as sns
%matplotlib inline
```

```python
sns.pairplot(dataset, x_vars=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'TAX', 'PTRATIO', 'B' ], y_vars='LSTAT', size=7, aspect=0.7, kind='reg')
```

```python
print("RESPOSTA NUM. 6")
feature_cols = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'TAX', 'PTRATIO', 'B']
x = dataset[feature_cols]
x.head()
```

```python
print("RESPOSTA NUM. 7")
print(type(x))
print(x.shape)
```

```python
print("RESPOSTA NUM. 6")
y = dataset['LSTAT']
y = dataset.LSTAT
y.head()
```

```python
print("RESPOSTA NUM. 7")
print(type(y))
print(y.shape)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
print("RESPOSTA NUM. 9")
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
print("RESPOSTA NUM. 11")
linreg = LinearRegression()
linreg.fit(x_train, y_train)
```

```python
print("RESPOSTA NUM. 12")
print(linreg.intercept_)
print(linreg.coef_)
```

```python
y_pred = linreg.predict(x_test)
```

```python
print("RESPOSTA NUM. 13")
list(zip(feature_cols, linreg.coef_))
```

```python
from sklearn import metrics
import numpy as np
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
